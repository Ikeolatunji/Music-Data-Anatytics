```python
def load_data(path,field_order, data):
    
      # Read and process the file
    with open(path, 'r',  encoding='utf-8') as file:
        header = next(file)
        for line in file:
            # Split the line by commas, handling quotes for the artists and name fields
            parts = []
            buffer = ""
            inside_quotes = False
            for char in line.strip():
                if char == '"':
                    inside_quotes = not inside_quotes
                elif char == ',' and not inside_quotes:
                    parts.append(buffer)
                    buffer = ""
                else:
                    buffer += char
            parts.append(buffer)  # Append the last part
            # print(parts)
                # Map the parts to their respective fields
            for i, field in enumerate(field_order):
                value = parts[i]
        
                    # Convert types where necessary
                if field in ['tempo','','valence','popularity', 'valence', 'acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'speechiness', 'acousticness', 'danceability','duration_ms',  'instrumentalness', 'liveness', 'loudness']:
                    value = float(value)
                elif field in ['year', 'duration_ms', 'key', 'explicit', 'mode']:
                    value = int(value)
                elif field == 'artists' or field == 'name' or field == 'genres':
                    value = value.strip('"').strip("[]").split(', ')
                    value = [v.strip("'") for v in value]
                elif field == 'loudness':
                    value = float(value)
                elif field == 'release_date':
                    value = value.strip('"')
        
                # Append the value to the appropriate list in the dictionary
                data[field].append(value)
        
                # Print the result to verify
    
    return data
           

field_order = ['valence', 'year', 'acousticness', 'artists', 'danceability', 'duration_ms','energy',
                  'explicit', 'id','instrumentalness', 'key', 'liveness', 'loudness', 'mode', 'name',
                  'popularity', 'release_date', 'speechiness', 'tempo']

data = {
        'valence': [],
        'year': [],
        'acousticness': [],
        'artists': [],
        'danceability': [],
        'duration_ms': [],
        'energy': [],
        'explicit': [],
        'id': [],
        'instrumentalness': [],
        'key':[],
        'liveness':[],
        'loudness': [],
        'mode': [],
        'name':[],
        'popularity':[], 
        'release_date': [],
        'speechiness':[],
        'tempo':[]
    }

data_genres = {
        'valence': [],
        'acousticness': [],
        'danceability': [],
        'duration_ms': [],
        'energy': [],
        'explicit': [],
        'id': [],
        'instrumentalness': [],
        'key':[],
        'liveness':[],
        'loudness': [],
        'mode': [],
        'name':[],
        'popularity':[], 
        'release_date': [], 
        'speechiness':[],
        'tempo':[],
        'genres':[]
    }

field_order2 = ['mode','genres','acousticness','danceability','duration_ms','energy','instrumentalness','liveness','loudness','speechiness','tempo','valence','popularity','key']

take_data = load_data('./data.csv', field_order, data)
print(take_data['artists'][15])

genre = load_data('./data_genres.csv', field_order2,data_genres)

print(genre['genres'][0])
```

```python
def mean(values):
    return sum(values) / len(values) if values else None

def mode(values):
    # Count frequency of each value
    frequency = {}
    for value in values:
        frequency[value] = frequency.get(value, 0) + 1
    max_freq = max(frequency.values())
    return [key for key, value in frequency.items() if value == max_freq]

def variance(values):
    m = mean(values)
    return sum((x - m) ** 2 for x in values) / len(values) if values else None

def standard_deviation(values):
    return variance(values) ** 0.5 if values else None

def minimum(values):
    return min(values) if values else None

def maximum(values):
    return max(values) if values else None

def describe(data, field):
    values = data.get(field, [])
    return {
        'mean': mean(values),
        'mode': mode(values),
        'variance': variance(values),
        'standard_deviation': standard_deviation(values),
        'min': minimum(values),
        'max': maximum(values)
    }
```

```python
def query_best_liveliness(data, genre_data):
    # Find the genre with the highest liveliness
    genre_liveliness = {}
    for i, genre in enumerate(genre_data['genres']):
        liveliness = data['liveness'][i]
        if genre:
            for g in genre:
                if g not in genre_liveliness:
                    genre_liveliness[g] = []
                genre_liveliness[g].append(liveliness)
    
    # Calculate the mean liveliness for each genre
    genre_mean_liveliness = {g: mean(liveliness) for g, liveliness in genre_liveliness.items()}
    
    # Get the genre with the best liveliness
    best_genre = max(genre_mean_liveliness, key=genre_mean_liveliness.get)
    return best_genre, genre_mean_liveliness[best_genre]

def query_artist_loudness(data, artist_data, query_type='highest'):
    # Query the artist with highest or lowest loudness
    artist_loudness = {}
    for i, artist in enumerate(artist_data['artists']):
        loudness = data['loudness'][i]
        for a in artist:
            if a not in artist_loudness:
                artist_loudness[a] = []
            artist_loudness[a].append(loudness)

    # Calculate the mean loudness for each artist
    artist_mean_loudness = {a: mean(loudness) for a, loudness in artist_loudness.items()}
    
    if query_type == 'highest':
        artist = max(artist_mean_loudness, key=artist_mean_loudness.get)
        loudness = artist_mean_loudness[artist]
    else:  # Query for lowest
        artist = min(artist_mean_loudness, key=artist_mean_loudness.get)
        loudness = artist_mean_loudness[artist]
    
    return artist, loudness
```
